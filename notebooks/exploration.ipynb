```python
# ===========================
# exploration.ipynb
# ===========================

# 1️ Import des librairies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------
# 2️ Lire le dataset
# ---------------------------
data = pd.read_csv("data/mental_health_lifestyle.csv")
print("Aperçu des premières lignes :")
display(data.head())

# ---------------------------
# 3️ Vérifier les valeurs manquantes
# ---------------------------
print("Valeurs manquantes par colonne :")
print(data.isnull().sum())

# Si besoin, supprimer les lignes vides
data = data.dropna()
print("Après suppression des lignes vides, forme du dataset :", data.shape)

# ---------------------------
# 4 Statistiques descriptives
# ---------------------------
print("Statistiques descriptives :")
display(data.describe())

# ---------------------------
# 5 Corrélations
# ---------------------------
plt.figure(figsize=(10,8))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Corrélation entre les variables")
plt.show()

# ---------------------------
# 6️ Visualisations individuelles
# ---------------------------

# Stress vs Dépression
plt.figure(figsize=(6,4))
sns.boxplot(x='Depression', y='Stress Level', data=data)
plt.title("Stress Level vs Dépression")
plt.show()

# Sommeil vs Dépression
plt.figure(figsize=(6,4))
sns.boxplot(x='Depression', y='Sleep Duration', data=data)
plt.title("Sleep Duration vs Dépression")
plt.show()

# Activité physique vs Dépression
plt.figure(figsize=(6,4))
sns.boxplot(x='Depression', y='Physical Activity', data=data)
plt.title("Physical Activity vs Dépression")
plt.show()

# Temps sur les réseaux sociaux vs Dépression
plt.figure(figsize=(6,4))
sns.boxplot(x='Depression', y='Social Media Usage', data=data)
plt.title("Social Media Usage vs Dépression")
plt.show()

# Qualité de l'alimentation vs Dépression
plt.figure(figsize=(6,4))
sns.boxplot(x='Depression', y='Diet Quality', data=data)
plt.title("Diet Quality vs Dépression")
plt.show()

# ---------------------------
# 7 Identifier les features importantes
# ---------------------------
# On peut se baser sur la corrélation avec 'Depression'
corr_dep = corr['Depression'].sort_values(ascending=False)
print("Corrélation des features avec Depression :")
display(corr_dep)
```
