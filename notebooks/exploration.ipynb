```python
# ===========================
# exploration.ipynb (VERSION CORRIG√âE)
# Notebook pour analyse exploratoire et visualisation
# ===========================

# 1Ô∏è‚É£ Import des librairies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Configuration des graphiques
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

# ---------------------------
# 2Ô∏è‚É£ V√©rifier le chemin et lire le dataset
# ---------------------------
print("üìÇ R√©pertoire actuel :", os.getcwd())
print("üìÅ Contenu du r√©pertoire :")
print(os.listdir())

# Ajuster le chemin selon o√π se trouve le notebook
try:
    # Si le notebook est dans notebooks/
    data = pd.read_csv("../data/mental_health_lifestyle.csv")
except FileNotFoundError:
    try:
        # Si le notebook est √† la racine
        data = pd.read_csv("data/mental_health_lifestyle.csv")
    except FileNotFoundError:
        print("‚ùå Fichier non trouv√©. V√©rifiez que le dataset est bien dans le dossier data/")
        print("üí° Ex√©cutez download_dataset.py pour t√©l√©charger le dataset")
        raise

print("‚úÖ Dataset charg√© avec succ√®s!")
print(f"üìä Dimensions : {data.shape[0]} lignes, {data.shape[1]} colonnes\n")

# ---------------------------
# 3Ô∏è‚É£ Aper√ßu des donn√©es
# ---------------------------
print("=" * 80)
print("üìã APER√áU DES DONN√âES")
print("=" * 80)
display(data.head(10))

print("\n" + "=" * 80)
print("üìä INFORMATIONS SUR LE DATASET")
print("=" * 80)
data.info()

print("\n" + "=" * 80)
print("üìà TYPES DE DONN√âES")
print("=" * 80)
print(data.dtypes)

# ---------------------------
# 4Ô∏è‚É£ V√©rifier les valeurs manquantes
# ---------------------------
print("\n" + "=" * 80)
print("üîç VALEURS MANQUANTES")
print("=" * 80)
missing = data.isnull().sum()
missing_pct = (missing / len(data)) * 100
missing_df = pd.DataFrame({
    'Colonne': missing.index,
    'Valeurs manquantes': missing.values,
    'Pourcentage': missing_pct.values
})
display(missing_df[missing_df['Valeurs manquantes'] > 0])

# Supprimer les lignes avec valeurs manquantes
if data.isnull().sum().sum() > 0:
    print(f"\n‚ö†Ô∏è {data.isnull().sum().sum()} valeurs manquantes d√©tect√©es")
    data_clean = data.dropna()
    print(f"‚úÖ Apr√®s nettoyage : {data_clean.shape[0]} lignes (supprim√© {data.shape[0] - data_clean.shape[0]} lignes)")
    data = data_clean
else:
    print("‚úÖ Aucune valeur manquante!")

# ---------------------------
# 5Ô∏è‚É£ Statistiques descriptives
# ---------------------------
print("\n" + "=" * 80)
print("üìä STATISTIQUES DESCRIPTIVES")
print("=" * 80)
display(data.describe())

# Statistiques pour les variables cat√©gorielles
categorical_cols = data.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    print("\nüìù Variables cat√©gorielles :")
    for col in categorical_cols:
        print(f"\n{col} :")
        print(data[col].value_counts())

# ---------------------------
# 6Ô∏è‚É£ Identifier la variable cible
# ---------------------------
print("\n" + "=" * 80)
print("üéØ VARIABLE CIBLE")
print("=" * 80)

# Chercher la colonne de d√©pression (peut avoir diff√©rents noms)
target_candidates = ['Depression', 'depression', 'Depressed', 'depressed', 'depression_status']
target_col = None

for candidate in target_candidates:
    if candidate in data.columns:
        target_col = candidate
        break

if target_col:
    print(f"‚úÖ Variable cible trouv√©e : {target_col}")
    print(f"\nDistribution de {target_col} :")
    print(data[target_col].value_counts())

    # Visualisation de la distribution
    plt.figure(figsize=(8, 6))
    data[target_col].value_counts().plot(kind='bar', color=['green', 'red'])
    plt.title(f'Distribution de {target_col}')
    plt.xlabel(target_col)
    plt.ylabel('Nombre de cas')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("‚ö†Ô∏è Variable cible non trouv√©e. Colonnes disponibles :")
    print(data.columns.tolist())

# ---------------------------
# 7Ô∏è‚É£ Matrice de corr√©lation
# ---------------------------
print("\n" + "=" * 80)
print("üîó MATRICE DE CORR√âLATION")
print("=" * 80)

# S√©lectionner uniquement les colonnes num√©riques
numeric_cols = data.select_dtypes(include=[np.number]).columns
data_numeric = data[numeric_cols]

plt.figure(figsize=(14, 10))
corr = data_numeric.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap="coolwarm",
            center=0, square=True, linewidths=1)
plt.title("Matrice de corr√©lation", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Corr√©lation avec la variable cible
if target_col and target_col in data_numeric.columns:
    print(f"\nüìä Corr√©lation avec {target_col} :")
    corr_target = corr[target_col].sort_values(ascending=False)
    display(corr_target)

    # Visualisation
    plt.figure(figsize=(10, 8))
    corr_target.drop(target_col).plot(kind='barh', color='steelblue')
    plt.title(f'Corr√©lation des features avec {target_col}')
    plt.xlabel('Coefficient de corr√©lation')
    plt.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
    plt.tight_layout()
    plt.show()

# ---------------------------
# 8Ô∏è‚É£ Visualisations par feature
# ---------------------------
print("\n" + "=" * 80)
print("üìà VISUALISATIONS PAR FEATURE")
print("=" * 80)

if target_col:
    # Liste des features num√©riques √† visualiser
    features_to_plot = [col for col in numeric_cols if col != target_col]

    # Cr√©er des boxplots
    n_features = min(len(features_to_plot), 9)  # Limiter √† 9 graphiques
    n_cols = 3
    n_rows = (n_features + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_features > 1 else [axes]

    for idx, feature in enumerate(features_to_plot[:n_features]):
        sns.boxplot(x=target_col, y=feature, data=data, ax=axes[idx], palette="Set2")
        axes[idx].set_title(f'{feature} vs {target_col}')
        axes[idx].set_xlabel(target_col)
        axes[idx].set_ylabel(feature)

    # Supprimer les axes vides
    for idx in range(n_features, len(axes)):
        fig.delaxes(axes[idx])

    plt.tight_layout()
    plt.show()

    # Violinplots pour les features les plus importantes
    top_features = corr_target.drop(target_col).abs().nlargest(4).index.tolist()

    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()

    for idx, feature in enumerate(top_features):
        sns.violinplot(x=target_col, y=feature, data=data, ax=axes[idx], palette="muted")
        axes[idx].set_title(f'{feature} vs {target_col}', fontweight='bold')

    plt.tight_layout()
    plt.show()

# ---------------------------
# 9Ô∏è‚É£ Distribution des variables num√©riques
# ---------------------------
print("\n" + "=" * 80)
print("üìä DISTRIBUTION DES VARIABLES NUM√âRIQUES")
print("=" * 80)

n_features = min(len(numeric_cols), 9)
n_cols = 3
n_rows = (n_features + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
axes = axes.flatten() if n_features > 1 else [axes]

for idx, col in enumerate(numeric_cols[:n_features]):
    data[col].hist(bins=30, ax=axes[idx], color='skyblue', edgecolor='black')
    axes[idx].set_title(f'Distribution de {col}')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Fr√©quence')

# Supprimer les axes vides
for idx in range(n_features, len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.show()

# ---------------------------
# üîü R√©sum√© et recommandations
# ---------------------------
print("\n" + "=" * 80)
print("üìù R√âSUM√â DE L'ANALYSE EXPLORATOIRE")
print("=" * 80)

print(f"""
‚úÖ Dataset charg√© : {data.shape[0]} lignes, {data.shape[1]} colonnes
‚úÖ Valeurs manquantes trait√©es
‚úÖ {len(numeric_cols)} variables num√©riques identifi√©es
‚úÖ {len(categorical_cols)} variables cat√©gorielles identifi√©es
""")

if target_col and target_col in data.columns:
    print(f"‚úÖ Variable cible : {target_col}")
    print(f"   Distribution : {data[target_col].value_counts().to_dict()}")

    # Identifier les features les plus importantes
    top_5_features = corr_target.drop(target_col).abs().nlargest(5)
    print(f"\nüéØ Top 5 features corr√©l√©es avec {target_col} :")
    for feature, value in top_5_features.items():
        print(f"   - {feature}: {value:.3f}")

print("\nüí° PROCHAINES √âTAPES :")
print("   1. Ex√©cuter model_training.py pour entra√Æner le mod√®le")
print("   2. Lancer l'application Streamlit : streamlit run app.py")

print("\n" + "=" * 80)
print("‚úÖ ANALYSE EXPLORATOIRE TERMIN√âE !")
print("=" * 80)
```
